## **① PromptTemplate**

#### PromptTemplate 예제

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate  # PromptTemplate class 임포트
from langchain_core.output_parsers import StrOutputParser
llm = ChatOpenAI(
     model_name="gpt-4o", 
     #api_key= "sk- 실습시 발급한 키 입력“
)
prompt= (                                                           #  PromptTemplate.from_template 메소드를 활용하여 프롬프트 개체 생성 
    PromptTemplate.from_template(
        """
        너는 요리사야. 내가 가진 재료들을 갖고 만들 수 있는 요리를 {개수}추천하고,
        그 요리의 레시피를 설명해줘. 내가 가진 재료는 아래와 같아.
        <재료>
        {재료}
        """
        )
    )
output_parser = StrOutputParser()		              # Outparser 개체 생성
chain = prompt | llm | output_parser                           # chain 생성
print(chain.invoke({"개수":"3", "재료" : "사과,양파,햄"}))   # 매개변수에 딕셔너리 자료형으로 데이터 입력 
```

## **② Output Parser**

#### CommaSeparatedListOutputParser 예제 (사용자가 지정한 숫자만큼 대상을  CSV  형태로 출력) 

```python
from langchain_core.output_parsers import CommaSeparatedListOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI

output_parser = CommaSeparatedListOutputParser()		# 콤마로 구분된 리스트 출력 파서 초기화

format_instructions = output_parser.get_format_instructions()	# 출력 형식 지침 가져오기
prompt = PromptTemplate(				# 프롬프트 템플릿 설정
    template="List five {subject}.\n{format_instructions}",	  	# 주제에 대한 다섯 가지를 나열하라는 템플릿
    input_variables=["subject"], 				# 입력 변수로 'subject' 사용
    partial_variables={"format_instructions": format_instructions}, 	# 부분 변수로 형식 지침 사용
)

# ChatOpenAI 모델 초기화
model = ChatOpenAI(
     model_name="gpt-4o", 
     #api_key= "sk- 실습시 발급한 키 입력“
)

chain = prompt | model | output_parser		# 프롬프트, 모델, 출력 파서를 연결하여 체인 생성
chain.invoke({"subject": "공포 영화"})		# “공포 영화"에 대한 chain 호출 실행
```

## **④ Text Splitters**

#### 지정된 청크 길이를 유지할 수 있는 RecursiveCharacterTextSplitter

```python
from langchain_community.document_loaders import PyPDFLoader      # PyPDFLoader 불러오기
file = "../실습파일/이슈리포트_2024-09호.pdf"
loader = PyPDFLoader(file)                              # PDF파일 불러올 객체 PyPDFLoader 선언
pages = loader.load()                                      # PDF파일 로드 

from langchain_text_splitters import RecursiveCharacterTextSplitter   # RecursiveCharacterTextSplitter 불러오기
text_splitter = RecursiveCharacterTextSplitter(
    separators=["\nn","\n"," ",""],                         # separator 리스트 설정
    chunk_size=500,            # 텍스트 분할 글자수, 청크의 최대크기
    chunk_overlap=100,       # 청크간 겹치는 글자수로 맥락 유지에 필요
    length_function=len       # 텍스트의 길이를 계산하는 함수 지정
   
)
                                            
texts = text_splitter.split_documents(pages)  # 텍스트 분할
```

#### 임베딩 완료 후 생성 된 벡터를 벡터 DB에 저장 

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma

embeddings_model = OpenAIEmbeddings(model = 'text-embedding-3-small')      # 임베딩 객체 생성
loader = PyPDFLoader("../실습파일/2022년 버전 선재 카탈로그_국문_최종.pdf")  
pages = loader.load()

text_splitter = RecursiveCharacterTextSplitter(      #PDF 문서를 여러 청크로 분할
    chunk_size=500,
    chunk_overlap=100 )

texts = text_splitter.split_documents(pages)

#OpenAI 임베딩 모델로 청크들을 임베딩 하면서 벡터 DB에 저장
db = Chroma.from_documents(texts, embeddings_model)
```

#### 임베딩 DB에 저장된 문서 검색해보기

```python
query = "쾌삭강"
docs =db.similarity_search(query)		# 유사도 검색을 통해 db에서 query와 유사한 chunk 추출
print(docs[0])			# 첫번째 문서 출력,  총 문서수는 len(docs)로 확인
```

```python
# 데이터베이스를 하드디스크에 저장
db1 = Chroma.from_documents(docs, embeddings_model, persist_directory=".\database")

# 데이터베이스를 하드디스크에서 불러오기기
db2 = Chroma(persist_directory=".\database", embedding_function=embeddings_model )
result = db2.similarity_search("쾌삭강")
print(result[0])

```

## **⑥ retriever**

#### 벡터 DB 기반 retriever

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma

#임베딩 모델 API 호출
embeddings_model = OpenAIEmbeddings(model = 'text-embedding-3-small')      # 임베딩 객체 생성
loader = PyPDFLoader("../실습파일/2022년 PosMAC1.5 카탈로그_강건재산업용_국문_최종.pdf") # PDF 문서 로드
pages = loader.load()

text_splitter = RecursiveCharacterTextSplitter (chunk_size=500, chunk_overlap=100 )
texts = text_splitter.split_documents(pages)				       # 문서분할 

#OpenAI 임베딩 모델로 청크들을 임베딩 하면서 벡터 DB에 저장
db = Chroma.from_documents(texts,embeddings_model)

retriever = db.as_retriever()		# retriever 객체 생성
retriever.invoke("내식성")		# 내식성으로 문서 검색 (기본 4건 출력) 
```

```python
result_score = db.similarity_search_with_relevance_scores("posmac")  # 값이 높을 수록 유사성 높음
print(result_score[0][1], result_score[1][1], result_score[2][1], result_score[3][1])
```

## **⑦ Chain 구성**

```python
# 모듈 불러오기기
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_chroma import Chroma
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_core.prompts import PromptTemplate

# 헌법 PDF 파일 로드
loader = PyPDFLoader(r"../실습파일/대한민국헌법(헌법)(제00010호)(19880225).pdf")
pages = loader.load_and_split()

# PDF 파일을 500자 청크로 분할
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=0)
docs = text_splitter.split_documents(pages)

# ChromaDB에 청크들을 벡터 임베딩으로 저장
db = Chroma.from_documents(docs, OpenAIEmbeddings(model = 'text-embedding-3-small',
                                                  #api_key= "sk-"
                                                  ))

# retriever 생성
retriever = db.as_retriever()
# prompt 생성
prompt = PromptTemplate.from_template(
    """
     당신은 질문-답변 작업을 위한 어시스턴트입니다.
     다음에 제공된 맥락을 사용하여 질문에 답하세요.
     만약 답을 모른다면, 그냥 모른다고 말하세요.
     한국어로 답변하세요.
     # question:
       {question}
     # context:
       {context}
     # answer:  """
)

# LLM 생성
llm = ChatOpenAI(model_name="gpt-4o", 
                 # api_key= "sk- "
                 temperature=0)

# Chain 생성
chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)

# 문서에 대한 질의를 입력하고, 답변을 출력
question = "국회의원의 임기는 몇년인가?"
response = chain.invoke(question)
print(response)

```
