# **6장 Agent 개요**

## **사업 아이디어 도출 Agent 구현 (sequential)**

```python
from crewai import Agent, Task, Crew, Process

print("=== CrewAI 아이디어 토론기 ===")
user_problem = input("\n[사용자 입력] 문제 또는 아이디어를 입력하세요:\n>")

# 1. Agent 정의

# agent1 : 제품 비전 중심
agent1 = Agent(
    role="Product Visionary (Jobs-inspired)",
    goal="단순하고 강력한 제품 아이디어를 제시한다.",
    backstory=(
        "당신은 사용자 경험과 미니멀리즘을 중시한다. "
        "복잡함을 제거하고, 한 문장으로 설명 가능한 아이디어를 선호한다."
    ),
    verbose=True,
)

# agent2 : 현실적 개선
agent2 = Agent(
    role="Pragmatic Technologist (Gates-inspired)",
    goal="현실적이고 실행 가능한 방향으로 아이디어를 발전시킨다.",
    backstory=(
        "당신은 기술, 사업성, 운영 관점에서 문제를 본다. "
        "아이디어의 허점을 짚고 구체적인 개선안을 제시한다."
    ),
    verbose=True,
)

# agent3 : 파괴적 확장
agent3 = Agent(
    role="Moonshot Engineer (Musk-inspired)",
    goal="아이디어를 더 크게 확장하고 혁신적인 방향을 제시한다.",
    backstory=(
        "당신은 기존 전제를 깨고 10배 큰 임팩트를 추구한다. "
        "대담하지만 논리적인 접근을 선호한다."
    ),
    verbose=True,
)

# agent4 : 최종 보고서 정리
agent4 = Agent(
    role="Report Writer",
    goal="토론 결과를 하나의 명확한 아이디어로 정리한다.",
    backstory=(
        "당신은 복잡한 논의를 구조화해 이해하기 쉬운 문서로 정리한다."
    ),
    verbose=True,
)

# 2. Task 정의
t1 = Task(
    description=(
        f"사용자 문제:\n{user_problem}\n\n"
        "제품 비전 관점에서 핵심 아이디어를 제시하라.\n"
        "- 한 줄 핵심\n"
        "- 간단한 설명\n"
        "- 차별점 3가지"
    ),
    expected_output="초기 아이디어",
    agent=agent1,
)

t2 = Task(
    description=(
        "이전 아이디어에 대해:\n"
        "- 문제점(반론)\n"
        "- 현실적인 개선안\n"
        "- 발전된 아이디어"
    ),
    expected_output="개선된 아이디어",
    agent=agent2, context=[t1],
)

t3 = Task(
    description=(
        "아이디어를 더 크게 발전시켜라.\n"
        "- 전제 깨기\n"
        "- 10배 확장\n"
        "- 리스크와 대응\n"
        "- 최종 아이디어"
    ),
    expected_output="확장된 아이디어",
    agent=agent3, context=[t1, t2],
)

t4 = Task(
    description=(
        "모든 토론을 종합해 최종 아이디어 보고서를 작성하라.\n\n"
        "[목차]\n"
        "- 아이디어명\n"
        "- 개요\n"
        "- 주요내용\n"
        "- 효과"
    ),
    expected_output="최종 아이디어 보고서",
    agent=agent4,
    context=[t1, t2, t3],
)

# 3. Crew 실행
print("\n[진행] 아이디어 토론을 시작합니다...\n")
crew = Crew(
    agents=[agent1, agent2, agent3, agent4],
    tasks=[t1, t2, t3, t4],
    process=Process.sequential,
    verbose=True,
)

result = crew.kickoff()

# 4. 결과 출력

print("\n====================")
print("최종 결과")
print("====================\n")
print(result)
```

## **사업 아이디어 도출 Agent 구현 (hierarchical)**

```python
from crewai import Agent, Task, Crew, Process
from crewai.tools import tool
from ddgs import DDGS

# 1. DuckDuckGo 검색 Tool
@tool("duckduckgo_search")
def duckduckgo_search(query: str) -> str:
    """
    DuckDuckGo로 웹 검색을 수행하고
    검색 결과의 제목, 요약, 링크를 반환한다.
    """
    results = []
    with DDGS() as ddgs:
        for r in ddgs.text(query, max_results=5):
            results.append(
                f"- {r.get('title')}\n"
                f"  {r.get('body')}\n"
                f"  {r.get('href')}"
            )
    return "\n".join(results) if results else "검색 결과 없음"


# 2. Agent 정의 
manager = Agent(
    role="Manager",
    goal="전체 작업을 계획하고 적절한 에이전트에게 Task를 위임한다.",
    backstory="프로젝트 전체 흐름을 조율하는 팀 리더.",  
    verbose=True,
)

researcher = Agent(
    role="Researcher",
    goal="주제와 관련된 최신 트렌드와 문제를 조사한다.",
    backstory="웹 검색과 정보 요약에 능숙한 리서처.",  
    tools=[duckduckgo_search],
    verbose=True,
)

ideator = Agent(
    role="Ideator",
    goal="조사 결과를 바탕으로 서비스 아이디어를 제안한다.",
    backstory="사용자 문제를 아이디어로 전환하는 기획자.",  
    verbose=True,
)


# 3. Task
topic = "2026년 대한민국 주요 트렌드에 기반한 새로운 사업모델"

research_task = Task(
    description=(
        f"주제: {topic}\n"
        "duckduckgo_search를 사용해 최근 트렌드나 문제를 3개 조사하고,\n"
        "각 항목을 한 줄 요약과 링크로 정리해줘."
    ),
    expected_output="트렌드 3개 목록(요약+링크)",
)

ideation_task = Task(
    description=(
        f"주제: {topic}\n"
        "조사 결과를 바탕으로 서비스 아이디어 1개를 한국어로 제안해줘.\n"
        "각 아이디어는 컨셉, 타깃, 핵심 기능 1개를 포함해야 한다."
    ),
    expected_output="아이디어 1개",
)


# 4. Crew
crew = Crew(
    agents=[researcher, ideator],
    tasks=[research_task, ideation_task],
    process=Process.hierarchical,
    manager_agent=manager,
    verbose=True,
)

# 5. 실행
result = crew.kickoff()
print("\n================ 최종 결과 ================\n")
print(result)


```
