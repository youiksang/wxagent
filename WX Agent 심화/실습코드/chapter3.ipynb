###  llm 객체 생성하여 간단한 질의 하기

```python
from langchain_openai import ChatOpenAI
# import httpx  사내 사용시 
# 객체 생성
llm = ChatOpenAI(
      # api_key= "sk- 실습시 발급한 키 입력"
      # http_client = httpx.Client(verify=False) 사내에서 사용시 입력
)
question = "대한민국의 수도는 어디인가요?"

# 질의
llm.invoke(question)
```

### 답변을 실시간으로 출력하기

```python
answer = llm.stream("부산의 아름다운 관광지 5곳과 꼭 맛보아야할 음식 5곳을 추천해줘")
# 스트리밍 방식으로 각 토큰을 출력 (실시간 출력)
for token in answer:
    print(token.content, end="", flush=True)
```

### llm의 출력을 구조화된 형태로 변환하는 출력파서(Output Parser) 

```python
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate

# 객체 생성
llm = ChatOpenAI(
     model_name="gpt-4o", 
     max_tokens= 100,
     #api_key= "sk- 실습시 발급한 키 입력" 
)

prompt = PromptTemplate.from_template("{topic}에 대해 쉽게 설명해줘")
output_parser = StrOutputParser()
chain = prompt | llm | output_parser    # chain 연결
input = {"topic" : "2차전지"}
chain.invoke(input)
```

### 프롬프트를 활용하여 영어 회화 출력하기

```python
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate

# 객체 생성
model = ChatOpenAI(
     model_name="gpt-4o", 
     #api_key= "sk- 실습시 발급한 키 입력"
)

template = """
당신은 영어 선생님입니다. 요청에 맞춰 [format]에 영어 회화를 작성해 주세요.

요청:
{question}

format:
- 영어 회화:
- 한글 해석:
"""

prompt = PromptTemplate.from_template(template)    # 프롬프트 템플릿을 이용하여 프롬프트 생성

output_parser = StrOutputParser()    # 문자열 출력 파서를 초기화

chain = prompt | model | output_parser # 체인 구성

scripts = chain.invoke({"question": "비행기표를 예매하고 싶어요"})  # 출력

print(scripts)
```
